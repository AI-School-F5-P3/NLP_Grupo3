```python
import numpy as np
import pandas as pd
import scipy.sparse
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
import lightgbm as lgb
from sklearn.model_selection import cross_val_score
import json
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

class ToxicCommentClassifier:
    def __init__(self, data_dir='processed_data'):
        """Initialize the classifier with data loading capabilities."""
        self.data_dir = Path(data_dir)
        self.models = {}
        self.results = {}
        
    def load_data(self):
        """Load the preprocessed and balanced data."""
        print("Loading preprocessed data...")
        
        # Load metadata to get file information
        with open(self.data_dir / 'metadata.json', 'r') as f:
            self.metadata = json.load(f)
            
        # Load features
        self.X_train = scipy.sparse.load_npz(self.data_dir / 'X_train_resampled.npz')
        self.X_test = scipy.sparse.load_npz(self.data_dir / 'X_test.npz')
        
        # Load labels
        self.y_train = np.load(self.data_dir / 'y_train_resampled.npy')
        self.y_test = np.load(self.data_dir / 'y_test.npy')
        
        print("Data loaded successfully!")
        print(f"Training set shape: {self.X_train.shape}")
        print(f"Test set shape: {self.X_test.shape}")
        print("\nClass distribution in training set:")
        print(pd.Series(self.y_train).value_counts(normalize=True))
        
    def setup_models(self):
        """Initialize different models for experimentation."""
        self.models = {
            'logistic': LogisticRegression(
                max_iter=1000,
                class_weight='balanced',
                random_state=42
            ),
            'svm': LinearSVC(
                max_iter=1000,
                class_weight='balanced',
                random_state=42
            ),
            'rf': RandomForestClassifier(
                n_estimators=100,
                class_weight='balanced',
                random_state=42,
                n_jobs=-1
            ),
            'lgb': lgb.LGBMClassifier(
                n_estimators=100,
                class_weight='balanced',
                random_state=42,
                n_jobs=-1
            )
        }
        
    def train_and_evaluate(self):
        """Train and evaluate all models."""
        for name, model in self.models.items():
            print(f"\nTraining {name}...")
            
            # Train the model
            model.fit(self.X_train, self.y_train)
            
            # Make predictions
            y_pred = model.predict(self.X_test)
            
            # Store results
            self.results[name] = {
                'classification_report': classification_report(self.y_test, y_pred, output_dict=True),
                'confusion_matrix': confusion_matrix(self.y_test, y_pred),
                'cross_val_scores': cross_val_score(
                    model, self.X_train, self.y_train, 
                    cv=5, scoring='f1', n_jobs=-1
                )
            }
            
            # Print results
            print(f"\nResults for {name}:")
            print(classification_report(self.y_test, y_pred))
            print(f"Cross-validation F1 scores: {self.results[name]['cross_val_scores']}")
            print(f"Mean CV F1: {self.results[name]['cross_val_scores'].mean():.3f} "
                  f"(±{self.results[name]['cross_val_scores'].std()*2:.3f})")
            
    def plot_results(self):
        """Visualize the results of all models."""
        # Create figure with subplots
        fig, axes = plt.subplots(2, 2, figsize=(15, 12))
        fig.suptitle('Model Performance Comparison')
        
        # Flatten axes for easier iteration
        axes = axes.flatten()
        
        for idx, (name, result) in enumerate(self.results.items()):
            # Plot confusion matrix
            sns.heatmap(
                result['confusion_matrix'],
                annot=True,
                fmt='d',
                cmap='Blues',
                ax=axes[idx]
            )
            axes[idx].set_title(f'{name} Confusion Matrix')
            axes[idx].set_xlabel('Predicted')
            axes[idx].set_ylabel('Actual')
            
        plt.tight_layout()
        plt.show()
        
        # Plot model comparison
        metrics = ['precision', 'recall', 'f1-score']
        comparison_data = {
            name: [result['classification_report']['weighted avg'][metric] 
                  for metric in metrics]
            for name, result in self.results.items()
        }
        
        comparison_df = pd.DataFrame(comparison_data, index=metrics)
        
        plt.figure(figsize=(10, 6))
        comparison_df.plot(kind='bar')
        plt.title('Model Performance Comparison')
        plt.ylabel('Score')
        plt.xticks(rotation=45)
        plt.legend(title='Models', bbox_to_anchor=(1.05, 1))
        plt.tight_layout()
        plt.show()
        
    def get_best_model(self):
        """Return the best performing model based on F1 score."""
        f1_scores = {
            name: result['classification_report']['weighted avg']['f1-score']
            for name, result in self.results.items()
        }
        best_model_name = max(f1_scores.items(), key=lambda x: x[1])[0]
        return best_model_name, self.models[best_model_name]

# Usage example
def run_model_experiments(data_dir='processed_data'):
    """Run all model experiments and return the best model."""
    
    # Initialize classifier
    classifier = ToxicCommentClassifier(data_dir)
    
    # Load data
    classifier.load_data()
    
    # Setup and train models
    classifier.setup_models()
    classifier.train_and_evaluate()
    
    # Visualize results
    classifier.plot_results()
    
    # Get best model
    best_model_name, best_model = classifier.get_best_model()
    print(f"\nBest performing model: {best_model_name}")
    
    return classifier

# Example usage in model_experiments.ipynb
if __name__ == "__main__":
    # Run experiments
    classifier = run_model_experiments()
```

```python
import numpy as np
import pandas as pd
import scipy.sparse
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_curve, roc_curve, auc
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
import lightgbm as lgb
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.calibration import CalibratedClassifierCV
import json
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from imblearn.pipeline import Pipeline
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import StandardScaler

class BalancedToxicClassifier:
    def __init__(self, data_dir='processed_data'):
        self.data_dir = Path(data_dir)
        self.models = {}
        self.results = {}
        self.best_thresholds = {}
        
    def load_data(self):
        """Load the preprocessed data."""
        print("Loading preprocessed data...")
        
        # Load features and labels
        self.X_train = scipy.sparse.load_npz(self.data_dir / 'X_train_resampled.npz')
        self.X_test = scipy.sparse.load_npz(self.data_dir / 'X_test.npz')
        self.y_train = np.load(self.data_dir / 'y_train_resampled.npy')
        self.y_test = np.load(self.data_dir / 'y_test.npy')
        
        # Print class distributions
        print("\nClass distribution in training set:")
        print(pd.Series(self.y_train).value_counts(normalize=True))
        print("\nClass distribution in test set:")
        print(pd.Series(self.y_test).value_counts(normalize=True))
        
    def setup_models(self):
        """Initialize models with balanced settings."""
        # Define class weights
        class_weights = dict(zip(
            np.unique(self.y_train),
            len(self.y_train) / (len(np.unique(self.y_train)) * np.bincount(self.y_train))
        ))
        
        self.models = {
            'logistic': Pipeline([
                ('scaler', StandardScaler(with_mean=False)),
                ('clf', LogisticRegression(
                    class_weight=class_weights,
                    max_iter=1000,
                    random_state=42
                ))
            ]),
            
            'svm': Pipeline([
                ('scaler', StandardScaler(with_mean=False)),
                ('clf', CalibratedClassifierCV(
                    LinearSVC(
                        class_weight=class_weights,
                        max_iter=1000,
                        random_state=42
                    ),
                    cv=5
                ))
            ]),
            
            'rf': Pipeline([
                ('scaler', StandardScaler(with_mean=False)),
                ('clf', RandomForestClassifier(
                    class_weight=class_weights,
                    n_estimators=200,
                    max_depth=20,
                    min_samples_leaf=5,
                    random_state=42,
                    n_jobs=-1
                ))
            ]),
            
            'lgb': Pipeline([
                ('scaler', StandardScaler(with_mean=False)),
                ('clf', lgb.LGBMClassifier(
                    class_weight=class_weights,
                    n_estimators=200,
                    num_leaves=31,
                    max_depth=20,
                    learning_rate=0.1,
                    random_state=42,
                    n_jobs=-1
                ))
            ])
        }
    
    def find_optimal_threshold(self, y_true, y_pred_proba):
        """Find the optimal decision threshold using precision-recall curve."""
        precisions, recalls, thresholds = precision_recall_curve(y_true, y_pred_proba[:, 1])
        f1_scores = 2 * (precisions * recalls) / (precisions + recalls + 1e-10)
        optimal_threshold = thresholds[np.argmax(f1_scores)]
        return optimal_threshold
    
    def evaluate_model(self, name, model, X, y_true):
        """Evaluate a single model with optimal threshold."""
        # Get probability predictions
        if hasattr(model, 'predict_proba'):
            y_pred_proba = model.predict_proba(X)
        else:
            # For models without predict_proba, use decision function
            decision_scores = model.decision_function(X)
            y_pred_proba = np.vstack([1-decision_scores, decision_scores]).T
            
        # Find optimal threshold
        optimal_threshold = self.find_optimal_threshold(y_true, y_pred_proba)
        self.best_thresholds[name] = optimal_threshold
        
        # Make predictions using optimal threshold
        y_pred = (y_pred_proba[:, 1] >= optimal_threshold).astype(int)
        
        # Calculate metrics
        results = {
            'classification_report': classification_report(y_true, y_pred, output_dict=True),
            'confusion_matrix': confusion_matrix(y_true, y_pred),
            'optimal_threshold': optimal_threshold,
            'roc_auc': roc_curve(y_true, y_pred_proba[:, 1]),
            'pr_curve': precision_recall_curve(y_true, y_pred_proba[:, 1])
        }
        
        return results
    
    def train_and_evaluate(self):
        """Train and evaluate all models with cross-validation."""
        skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
        
        for name, model in self.models.items():
            print(f"\nTraining {name}...")
            
            # Train the model
            model.fit(self.X_train, self.y_train)
            
            # Evaluate on test set
            self.results[name] = self.evaluate_model(
                name, model, self.X_test, self.y_test
            )
            
            # Print results
            print(f"\nResults for {name} (threshold = {self.best_thresholds[name]:.3f}):")
            print(classification_report(
                self.y_test,
                (model.predict_proba(self.X_test)[:, 1] >= self.best_thresholds[name]).astype(int)
            ))
    
    def plot_results(self):
        """Visualize model performance with ROC and PR curves."""
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
        
        # Plot ROC curves
        for name, result in self.results.items():
            fpr, tpr, _ = result['roc_auc']
            roc_auc = auc(fpr, tpr)
            ax1.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc:.2f})')
            
        ax1.plot([0, 1], [0, 1], 'k--')
        ax1.set_xlabel('False Positive Rate')
        ax1.set_ylabel('True Positive Rate')
        ax1.set_title('ROC Curves')
        ax1.legend()
        
        # Plot PR curves
        for name, result in self.results.items():
            precision, recall, _ = result['pr_curve']
            ax2.plot(recall, precision, label=name)
            
        ax2.set_xlabel('Recall')
        ax2.set_ylabel('Precision')
        ax2.set_title('Precision-Recall Curves')
        ax2.legend()
        
        plt.tight_layout()
        plt.show()
        
        # Plot confusion matrices
        n_models = len(self.results)
        fig, axes = plt.subplots(2, 2, figsize=(15, 12))
        axes = axes.flatten()
        
        for idx, (name, result) in enumerate(self.results.items()):
            sns.heatmap(
                result['confusion_matrix'],
                annot=True,
                fmt='d',
                cmap='Blues',
                ax=axes[idx]
            )
            axes[idx].set_title(f'{name} Confusion Matrix\n(threshold = {self.best_thresholds[name]:.3f})')
            
        plt.tight_layout()
        plt.show()

def run_balanced_experiments(data_dir='processed_data'):
    """Run experiments with balanced evaluation."""
    classifier = BalancedToxicClassifier(data_dir)
    classifier.load_data()
    classifier.setup_models()
    classifier.train_and_evaluate()
    classifier.plot_results()
    return classifier

# Example usage
if __name__ == "__main__":
    classifier = run_balanced_experiments()
```
